**4.	Count the number of shows released per year (release_year).**

**9.	List all shows where "Samuel Jouy" is part of the cast.**

```python
samuel_jouy_shows = df[df['cast'].str.contains('Samuel Jouy', na=False)]
samuel_jouy_shows
```

**10.Find all movies with a duration of more than 90 minutes (ensure the duration column is parsed correctly for numerical comparisons).**

```python
movies = df[df['type'] == 'Movie']
movies['duration_minutes'] = movies['duration'].str.extract('(\d+)').fillna(0).astype(int)
movies_more_than_90 = movies[movies['duration_minutes'] > 90]
movies_more_than_90
```

**11. Find all shows that belong to both "Crime TV Shows" and "International TV Shows."**

```python
crime_international_shows = df[df['listed_in'].str.contains('Crime TV Shows') & df['listed_in'].str.contains('International TV Shows')]
crime_international_shows
```

**12.Find all TV shows added after January 1, 2020, that were released in India.**

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
india_shows_2020 = df[(df['date_added'] > '2020-01-01') & (df['country'] == 'India') & (df['type'] == 'TV Show')]
india_shows_2020
```
